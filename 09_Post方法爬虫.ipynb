# Post方法爬虫

# 1. Post方法爬虫

① post方法要加页面的data数据。

```python
import requests
from lxml import etree
url = 'https://rsdata.chinaventure.com.cn/api/getNewsListByChannel'
# post方法要加页面的data数据
data = {
        "channel_id": 113,
        "begin_num": 21,  # post方法的第一条数据为21条新闻
        "size": 60,
        "page": 1
        }

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36"}      
n = 1
while True:
    res = requests.post(url,data=data,headers = headers)
    print(len(res.json()["data"]["channel_news_list"]))
    news = res.json()["data"]["channel_news_list"]
    if not news:  
        break
    for new in news:
        print(str(n),new["news_title"])
        n+=1
    data["begin_num"]=data["begin_num"]+data["size"]
    
# 前20条新闻是get方法
print("下面是get获得的新闻数据")
get_url = 'https://www.chinaventure.com.cn/news/113.html'
get_res = requests.get(get_url,headers=headers)
print(get_res)
get_res.encoding='uft-8'
html = etree.HTML(get_res.text)
news = html.xpath('//*[@id="page_content"]/div[4]/div[1]/div[2]/div[1]/ul/li'  )
for new in news:
    print(str(n),new.xpath('./a/h1/text()')[0])
    n+=1
```
