## 2.1 MySQL创建表语句

# 3. Python连接MySQL

## 3.1 pymysql指定数据库

```python
#已经配准MySQL了，当前的MySQL运行在本地，用户名为root，密码为password，运行端口为3306.
#这里利用 PyMySQL 先连接 PyMySQL，然后创建一个新的数据库，名字叫做spiders

# 连接mysql数据库
import pymysql
# host='127.0.0.1' 表示本地，相当于 host = 'localhost'
mysqldb = pymysql.connect(host='127.0.0.1',
                         user = 'root',
                         password = 'password',
                         database = 'ai',
                         port = 3306)

# 获取游标
cursor = mysqldb.cursor()

# 数据库中语句
select_sql = '''
SELECT * from student t  # student 重命名为t
left join
# student1 重命名为t1
student1 t1 on t.num = t1.num where t.age>10 order by age desc
'''

# 通过游标执行查询语句
cursor.execute(select_sql)

# 通过游标获得所有查询结果
data = cursor.fetchall()
print(data)
cursor.close() # 游标用完通常要关闭一下
```

## 3.3 封装查询数据库语句

```python
#coding=utf-8
import pymysql
class PythonMysql(object):
    # 构造方法
    def __init__(self,host,user,password,port):
        self.host       =     host
        self.user       =     user
        self.password   =     password
        self.port       =     port
        self.cursor     =     self.connect_mysql()   # 初始化的时候调用了连接数据库语句
        
    # 连接数据库
    def connect_mysql(self):
        try:
            mysql_db = pymysql.connect(  host       =     self.host,
                                         user       =     self.user,
                                         password   =     self.password,
                                         port       =     self.port)
            return mysql_db.cursor()    # 返回的是一个游标
        
        except Exception as e:
            print(e)
        
    # 查询数据
    def select_mysql(self,qurey_str): # 传入查询语句
        self.cursor.execute(qurey_str)
        return self.cursor.fetchall()   # 返回查询的数据
        

# 数据库中语句
select_sql = '''
SELECT * from ai.student t  
left join
ai.student1 t1 on t.num = t1.num where t.age>10 order by age desc
'''

mysql_object = PythonMysql('localhost','root','password',3306)   # 实例化对象                 
if mysql_object.cursor:    # 如果为空就不运行
    data = mysql_object.select_mysql(select_sql)
    print(data)
```
